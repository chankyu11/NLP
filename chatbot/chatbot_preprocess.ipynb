```python
path = 'C:/nlp/tensorflow-ml-nlp-tf2/6.CHATBOT/data_in/ChatBotData.csv'
vocab_path ='C:/nlp/tensorflow-ml-nlp-tf2/6.CHATBOT/data_in/vocabulary.txt'
```

```python
inputs, outputs = load_data(path)
```

```python
word2idx, idx2word, vocab_size = load_vocabulary(path, vocab_path)
```

```python
index_inputs, input_seq_len = enc_processing(inputs, word2idx, tokenize_as_morph=True)
index_outputs, output_seq_len = dec_output_processing(outputs, word2idx, tokenize_as_morph=True)
index_targets = dec_target_processing(outputs, word2idx, tokenize_as_morph=True)
```

```python
data_configs = {}
data_configs['word2idx'] = word2idx
data_configs['idx2word'] = idx2word
data_configs['vocab_size'] = vocab_size
data_configs['pad_symbol'] = pad
data_configs['std_symbol'] = std
data_configs['end_symbol'] = end
data_configs['unk_symbol'] = unk
```

```python
data_path = 'C:/nlp/tensorflow-ml-nlp-tf2/6.CHATBOT/data_in/'
TRAIN_INPUTS = 'train_inputs.npy'
TRAIN_OUTPUTS = 'train_outputs.npy'
TRAIN_TARGETS = 'train_targets.npy'
DATA_CONFIGS = 'data_configs.json'

np.save(open(data_path + TRAIN_INPUTS, 'wb'), index_inputs)
np.save(open(data_path + TRAIN_OUTPUTS , 'wb'), index_outputs)
np.save(open(data_path + TRAIN_TARGETS , 'wb'), index_targets)

json.dump(data_configs, open(data_path + DATA_CONFIGS, 'w'))
```
